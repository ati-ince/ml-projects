````python
# The Melbourne data has some missing values (some houses for which some variables weren't recorded.)
# We'll learn to handle missing values in a later tutorial.
# Your Iowa data doesn't have missing values in the columns you use.
# So we will take the simplest option for now, and drop houses from our data.
# Don't worry about this much for now, though the code is:

# dropna drops missing values (think of na as "not available")
melbourne_data = melbourne_data.dropna(axis=0)
````

````python
# lets call for test piece of the data from DS
y = melbourne_data.Price
print(y)
````

**FEATURE**

According to my research, every ML model (e.g., classification model like Random Forest Classificator -aka RFC-) need to choose the most important features to work on (for training, classification, etc.)

````python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
````

````python
# to call featured data from main dataset
X = melbourne_data[melbourne_features]
X.head()
````

### BUILDING THE MODEL

Here is the building steps for the model:

1. **Define:** What type of model we going to build, like decision tree, or others
2. **Fit:** Capture pattern from providid data
3. **Predict**
4. **Avaluate**

````python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y) # fitting features to the expected (Price)
# -> this is CURVE FITTING, simplest prediction methodology
````

````python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head())) # it is predicting the PRICE(s)
````

### 4. MODEL VALIDATION
The last part, we use curvefitting technique (Decision Tree Regressor) and Fitting Features and Expected Value (Price). This section, we focuses validation (means use training and test divition using the data)

For calculation model quality (between predicted value and actual value) there are multiple metrics. We will use **Mean Absolute Error (MAE)** for starting.



```
error = actualValue - predictedValue
```



````python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
````

````python
# for validation, need to divide dataset training and testing section.
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0) # divide traijn and test groups

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_y)


val_predict = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predict))

````
